```python
import pandas as pd
```

```python
support_data_raw = pd.read_csv('supporting_master_data_raw.csv')
support_data_raw.head(2)
```

```python
application_test_clean = pd.read_csv('application_test_cleaned.csv')
application_test_clean.head(2)
```

```python
application_train_clean = pd.read_csv('application_train_cleaned.csv')
application_train_clean.head(2)
```

```python
# Merge the cleaned application_train with supporting master data
master_data_train = application_train_clean.merge(
    support_data_raw,
    on='SK_ID_CURR',
    how='left'
)

# Merge the cleaned application_test with supporting master data
master_data_test = application_test_clean.merge(
    support_data_raw,
    on='SK_ID_CURR',
    how='left'
)

print("Merged master_data_train shape:", master_data_train.shape)
print("Merged master_data_test shape:", master_data_test.shape)

# Save the merged datasets
master_data_train.to_csv("master_data_train.csv", index=False)
master_data_test.to_csv("master_data_test.csv", index=False)

print("Merged datasets saved successfully!")
```
